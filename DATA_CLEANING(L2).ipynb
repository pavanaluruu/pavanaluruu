```python
import numpy as np
data = [1,2,2,3,1,1,15,2,2,3,1,1,2]
mean = np.mean(data)
std = np.std(data)
print('mean of the dataset is',mean)
print('std.deviation is',std)
threshold = 3
outlier = []
for i  in data:
  z = (i-mean)/std
  if z > threshold:
    outlier.append(i)
  print('outkier in dataset is',outlier)
```

INTERQUARTILE RANGE TO DETECT OUTLIERS IN DATA

1)represents tye 25th percentile of the data.
2)50th percentile of the data.
3)75th
4)2n/2n+1 data points

IQR=Q3-Q1

```python
import numpy as np
import seaborn as sns
data=[6,2,3,4,1,5,50]
sort_data = np.sort(data)
sort_data
Q1 = np.percentile(data,25,interpolation = 'midpoint')
Q2 = np.percentile(data,50,interpolation = 'midpoint')
Q3 = np.percentile(data,75,interpolation = 'midpoint')
print('Q1 25 percentile of the given data is,',Q1)
print('Q2 50 percentile of the given data is,',Q2)
print('Q3 75 percentile of the given data is,',Q3)
IQR = Q3-Q1
print('Interquartile range is',IQR)
```

```python
low_lin = Q1 - 1.5*IQR
up_lin = Q3+1.5*IQR
print('low_limit is ',low_lin)
print('up_limit is',up_lin)
```
